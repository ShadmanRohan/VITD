```python
import fasttext
import fasttext.util

# Download the Bengali model (you can choose the dimension: 300M is common)
#fasttext.util.download_model('bn', if_exists='ignore')  # Bengali
model_path = './fasttext/cc.bn.300.bin'
ft = fasttext.load_model(model_path)

def sentence_vector(sentence):
    words = sentence.split()

    # Calculate the average vector representation of all words in the sentence
    vec = sum([ft.get_word_vector(word) for word in words]) / len(words)
    return vec
```
